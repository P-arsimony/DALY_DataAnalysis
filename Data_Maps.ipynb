# Mapas de Luminarias

## Librerías necesarias


```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

## Cargar datos

```python
# capa de entidad, en este caso solo la cdmx
mg_ent = gpd.read_file("data_raw/09_ciudaddemexico.zip!conjunto_de_datos", layer="09ent")
display( mg_ent )

# capa de entidad, en este caso las 16 alcaldias
mg_mun = gpd.read_file("data_raw/09_ciudaddemexico.zip!conjunto_de_datos", layer="09mun")
display( mg_mun )

# capa de entidad, en este caso las 16 alcaldias
mg_loc = gpd.read_file("data_raw/09_ciudaddemexico.zip!conjunto_de_datos", layer="09l")
display( mg_loc )

df = pd.read_csv('data_clean/cmdx_servicios_publicos.csv')
display(df)
```

```python
# Se cambia nombre de columna para hacer el merge.
df = df.rename(columns={'ubicageo_f': 'CVEGEO', 'descmpio': 'NOMGEO'})
display(df)
```

```python
# Diccionario de mapeo para corregir los códigos de CVGEO
CVGEO_codigos = {
    '9002': '09002',
    '9003': '09003',
    '9004': '09004',
    '9005': '09005',
    '9006': '09006',
    '9007': '09007',
    '9008': '09008',
    '9009': '09009',
    '9010': '09010',
    '9011': '09011',
    '9012': '09012',
    '9013': '09013',
    '9014': '09014',
    '9015': '09015',
    '9016': '09016',
    '9017': '09017'
}

# Se aplica el mapeo para la corrección de códigos.
df['CVEGEO'] = df['CVEGEO'].astype(str).map(CVGEO_codigos).fillna(df['CVEGEO'].astype(str))
```

```python
# Merge de datos geoespaciales con datos tabulares.

# Conversión de CVEGEO a str en ambos dataframes
df['CVEGEO'] = df['CVEGEO'].astype(str)
mg_mun['CVEGEO'] = mg_mun['CVEGEO'].astype(str)

# Now merge
df_loc = df.merge(mg_mun, on=['CVEGEO', 'NOMGEO'], how='left')
display(df_loc.head())
```

### Elaboración de mapas de luminarias por municipio.

```python
# Conversión a GeoDataFrame
df_loc = gpd.GeoDataFrame(df_loc, geometry='geometry')

# Chequeo del tipo de DataFrame y CRS
print(type(df_loc))  
print(df_loc.crs)    
```

```python
# Mapas de luminarias en funcionamiento por municipio de CDMX
fig, axes = plt.subplots(1, 2, figsize=(14, 10))

########################################
# Mapa de luminarias en funcionamiento #
########################################

df_loc.plot(
    column='serfun1',  
    cmap='viridis',
    edgecolor='black',
    legend=True,
    ax=axes[0])  

# Agregar nombres SOLO para municipios con más de X luminarias
umbral = df_loc['serfun1'].quantile(0.7) 

for idx, row in df_loc.iterrows():
    if row['serfun1'] > umbral:
        centroid = row['geometry'].centroid
        axes[0].text(centroid.x, centroid.y,  
                row['NOMGEO'], 
                fontsize=8, 
                ha='center',
                fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))

axes[0].set_title('Luminarias en funcionamiento')  


##############################################
# Mapa de luminarias fuera de funcionamiento #
##############################################

df_loc.plot(
    column='serfun2',  
    cmap='viridis',
    edgecolor='black',
    legend=True,
    ax=axes[1])

# Agregar nombres solo para municipios con más de X luminarias
umbral = df_loc['serfun2'].quantile(0.7) 

for idx, row in df_loc.iterrows():
    if row['serfun2'] > umbral:
        centroid = row['geometry'].centroid
        axes[1].text(centroid.x, centroid.y, 
                row['NOMGEO'], 
                fontsize=8, 
                ha='center',
                fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))
        
axes[1].set_title('Luminarias fuera de funcionamiento')  

plt.tight_layout()
plt.show()
```
