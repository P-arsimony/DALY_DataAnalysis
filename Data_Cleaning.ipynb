# Limpieza de datos-
A raíz de que los datos elegidos para el proyecto tienen problemas de multiíndice y filas con celdas vacías, se realizará una limpieza de datos para poder trabajar con ellos.

```python
# Cargamos datos y pandas.
import pandas as pd
filename = 'data_raw/DALYs/ghe2021_daly_global_new.xlsx'
df = pd.read_excel(filename)
df
```

Dado que el dataset tiene un multíindice, se usará el método ffill para rellenar las celdas vacías con el valor de la fila anterior. Esto permitirá que cada fila tenga toda la información necesaria para su análisis posterior.

Además, necestimos leer la hoja de excel de los años que nos interesan (2000-2021) y aplicar el método ffill a cada una de ellas para obtener un dataset limpio y listo para su análisis.

```python
# Lectura de hoja "Global 2021"
G2021= pd.read_excel(filename, sheet_name="Global 2021")
G2021
# Lectura de hoja "Global 2020"
G2020 = pd.read_excel(filename, sheet_name="Global 2020")
G2020
# Lectura de hoja "Global 2019"
G2019 = pd.read_excel(filename, sheet_name="Global 2019")
G2019
# Lectura de hoja "Global 2015"
G2015 = pd.read_excel(filename, sheet_name="Global 2015")
G2015
# Lectura de hoja "Global 2010"
G2010 = pd.read_excel(filename, sheet_name="Global 2010")
G2010
# Lectura de hoja "Global 2000"
G2000 = pd.read_excel(filename, sheet_name="Global 2000")
G2000
```

```python
# Usamos el método ffill para rellenar las celdas vacías con el valor de la fila anterior.
# Realizamos este proiceso a cada una delas hojas que leímos con aterioridad.

G2021 = G2021.ffill()
G2020 = G2020.ffill()
G2019 = G2019.ffill()
G2015 = G2015.ffill()
G2010 = G2010.ffill()
G2000 = G2000.ffill()
```

```python
# Guardamos cada año como un excel limpio en capeta data_clean.
G2021.to_excel('data_clean/DALY_ffill_2021.xlsx', index=False, sheet_name='Global 2021')
G2020.to_excel('data_clean/DALY_ffill_2020.xlsx', index=False, sheet_name='Global 2020')
G2019.to_excel('data_clean/DALY_ffill_2019.xlsx', index=False, sheet_name='Global 2019')
G2015.to_excel('data_clean/DALY_ffill_2015.xlsx', index=False, sheet_name='Global 2015')
G2010.to_excel('data_clean/DALY_ffill_2010.xlsx', index=False, sheet_name='Global 2010')
G2000.to_excel('data_clean/DALY_ffill_2000.xlsx', index=False, sheet_name='Global 2000')
```
