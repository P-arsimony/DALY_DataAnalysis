# Limpieza de datos-
A raíz de que los datos elegidos para el proyecto tienen problemas de multiíndice y filas con celdas vacías, se realizará una limpieza de datos para poder trabajar con ellos.

## Objetivo de la remodelación y limpieza de datos.

Lo que se observa en los conjuntos de datos que se generaron para cada año es que aún los nombres de las columnas se mantienen como unnamed. Lo que queremos conseguir es que cada columna sea una variable y cada obseración sea una fila. El resultado al que se quiere llegar es el siguiente:

| Nombre de la variables  | Descripción     |
| :-------------: | :------------- |
| categoria_principal | Categoría principal de la enfermedad |
| categoria_nivel1 | Subcategoría |
| categoria_nivel2 | Subsubcategoría (si la hay) |
| nombre_enfermedad | Nombre de la enfermedad. Es el nivel más bajo dentro de las categorías de la OMS.|
| codigo| Código según la OMS. |
| ano | Año correspondiente de la hoja. |
| sexo | Sexo de la persona afectada. |
| edad_min | Límite inferior del rango de edad.|
| edad_max | Límite superior del rango de edad.|
| dalys | Valor numérico de los DALYs para la enfermedad. |

**Tabla 1:** Objetivo de la remodelación y limpieza de datos. Se busca que cada columna sea una variable y cada fila una observación.

Por lo tanto, el siguiente paso será remodelar los conjuntos de datos por año para conseguir un formato adecuado para el análisis posterior. Para esto...

```python
# Librerías necesarias
import pandas as pd
import numpy as np
```

```python
# Definimos el path del archivo .xslsx y los años que nos interesan para el análisis.
archivo = 'data_raw/DALYs/ghe2021_daly_global_new.xlsx'
```

```python
# Cargamos datos y pandas.
df = pd.read_excel(archivo)
df
```

Dado que el dataset tiene un multíindice, se usará el método ffill para rellenar las celdas vacías con el valor de la fila anterior. Esto permitirá que cada fila tenga toda la información necesaria para su análisis posterior.

Además, necestimos leer la hoja de excel de los años que nos interesan (2000-2021) y aplicar el método ffill a cada una de ellas para obtener un dataset limpio y listo para su análisis.

```python
# Lectura de hoja "Global 2021"
G2021= pd.read_excel(archivo, sheet_name="Global 2021")
G2021
# Lectura de hoja "Global 2020"
G2020 = pd.read_excel(archivo, sheet_name="Global 2020")
G2020
# Lectura de hoja "Global 2019"
G2019 = pd.read_excel(archivo, sheet_name="Global 2019")
G2019
# Lectura de hoja "Global 2015"
G2015 = pd.read_excel(archivo, sheet_name="Global 2015")
G2015
# Lectura de hoja "Global 2010"
G2010 = pd.read_excel(archivo, sheet_name="Global 2010")
G2010
# Lectura de hoja "Global 2000"
G2000 = pd.read_excel(archivo, sheet_name="Global 2000")
G2000
```

```python
# Revisamos cómo se ven los datos de 2021 para identificar la filas que no sirven.
print(G2021.head(10))
```

```python
nombres_columnas = G2021.iloc[5].tolist()
print(nombres_columnas)
```

```python
# Dado que las primeras 6 filas no contienen datos útiles, las eliminamos y renombramos las columnas. Esto es, los encabezados están en la fila 5 (index 4)
G2021.columns = nombres_columnas
G2021 = G2021.iloc[6:].reset_index(drop=True)
print(G2021.head(5))
```

```python
# Aplicamos ffill a las primeras 4 columnas para rellenar los valores faltantes.
# Aplicamos ffill de una columna a la vez, ya que si lo hacemos de una sola vez, se rellenan con el valor de la columna anterior.

columnas_jerarquia = G2021.columns[:4].tolist()
for col in columnas_jerarquia:
    G2021[col] = G2021[col].ffill()
```

```python
# Dado que existen algunas columnas sin nombre, asignamos un nombre genérico para evitar problemas con el cambio de nombre.

nuevos_nombres = []
for i, col in enumerate(G2021.columns):
    if col == 'Causa':
        nuevos_nombres.append(f'Causa_{i}')
    else:
        nuevos_nombres.append(col if pd.notna(col) else f'col_{i}')

G2021.columns = nuevos_nombres
```

```python
# Cambiamos el nombre de las primeras 5 columnas.

G2021 = G2021.rename(columns={
    G2021.columns[0]: 'codigo',
    G2021.columns[1]: 'categoria_principal',
    G2021.columns[2]: 'categoria_nivel1',
    G2021.columns[3]: 'categoria_nivel2',
    G2021.columns[4]: 'causa'
})
```

```python
# Igualmente, dado que no nos interesan las columnas correspondientes al total de DALYs por grupo etario para mujeres, hombres y ambos sexos, las eliminamos.
columnas_eliminar = [6, 7, 8]
G2021 = G2021.drop(columns=G2021.columns[columnas_eliminar])
```

Un problema que tiene el conjunto es que hay una fila de población correspondiente a cada edad. Si hacemos primero `pd.melt()`, debemos seprar la fila de población y luego volverla a incorporar al dataset.

```python
# Separamos la fila de población del resto de los datos para luego volverla a incorporar al dataset.
poblacion = G2021['codigo'] == 'Population (thousands)'
poblacion_df = G2021[poblacion].copy()

# Copiamos el dataset sin la fila de población para trabajar con él.
enfermedades_df = G2021[~poblacion].copy()
```

```python
print(poblacion_df)
```

```python
print(enfermedades_df.head())
```

```python
columnas_edad = poblacion_df.columns[5:].tolist() 
```

```python
# Ahora identificamos 

poblacion_long = pd.melt(
    poblacion_df,
    id_vars=['codigo', 'categoria_principal', 'categoria_nivel1', 'categoria_nivel2', 'causa','Age group'],
    value_vars=columnas_edad,
    var_name='grupo_edad',
    value_name='poblacion'
)

poblacion_long['poblacion'] = pd.to_numeric(poblacion_long['poblacion'], errors='coerce')
```

Dado que el formato de los datos está en *wide*, se usará la función `pd.melt()` para transformar el dataset a formato *long*, lo que permitirá que las edades se conviertan en una variable y cada fila represente una observación única de enfermedad, sexo, edad y año.

```python
# Definimos variables de ID y variables de edad para usar en la función melt.

id_vars = ['codigo', 'categoria_principal', 'categoria_nivel1', 'categoria_nivel2', 'Causa', 'Age group']
edad_vars = [col for col in G2021.columns if col not in id_vars]
```

```python
G2021_long = pd.melt(
    G2021, 
    id_vars=id_vars, 
    value_vars=edad_vars, 
    var_name='Edad', 
    value_name='dalys'
)
```
